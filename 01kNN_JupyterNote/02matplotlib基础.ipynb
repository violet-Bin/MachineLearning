# matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = "--")
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = "--")
plt.xlim(-5, 15)
plt.ylim(0, 2)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = "--")
plt.axis([-2, 20, -4, 4])
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = "--")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.plot(x, siny, label = "sin(x)")
plt.plot(x, cosy, color = "red", linestyle = "--", label = "cos(x)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
```

```python
plt.plot(x, siny, label = "sin(x)")
plt.plot(x, cosy, color = "red", linestyle = "--", label = "cos(x)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("welcome to the ML World")
plt.legend()
```
